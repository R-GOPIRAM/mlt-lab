```python
import pandas as pd

# 1. Create a sample DataFrame
data = {
    'Name': ['John', 'Jane', 'Bob', 'Mary'],
    'Age': [30, 25, 40, 35],
    'Gender': ['M', 'F', 'M', 'F']
}
df = pd.DataFrame(data)

# 2. Print the full DataFrame
print("Original DataFrame:")
print(df)

# 3. Select specific columns
print("\nSelected Columns (Name & Age):")
print(df[['Name', 'Age']])

# 4. Filter rows where Age > 30
print("\nRows where Age > 30:")
print(df[df['Age'] > 30])

# 5. Add a new column 'Salary'
df['Salary'] = [50000, 60000, 70000, 80000]
print("\nDataFrame after adding 'Salary' column:")
print(df)

# 6. Drop the 'Gender' column
df = df.drop('Gender', axis=1)
print("\nDataFrame after dropping 'Gender' column:")
print(df)

# 7. Create a new DataFrame from list of dictionaries
new_data = [
    {'Name': 'Alice', 'Age': 25, 'Gender': 'Female'},
    {'Name': 'Bob', 'Age': 30, 'Gender': 'Male'},
    {'Name': 'Charlie', 'Age': 35, 'Gender': 'Male'},
    {'Name': 'David', 'Age': 40, 'Gender': 'Male'},
    {'Name': 'Eve', 'Age': 45, 'Gender': 'Female'}
]
new_df = pd.DataFrame(new_data)

# 8. Print the new DataFrame
print("\nNew DataFrame:")
print(new_df)

# 9. Group by 'Gender' and calculate average age
grouped = new_df.groupby('Gender')['Age'].mean()
print("\nAverage Age by Gender:")
print(grouped)
```
