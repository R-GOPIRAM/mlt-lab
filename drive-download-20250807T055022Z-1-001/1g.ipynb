```python
from scipy import linalg
import numpy as np  # define square matrix

two_d_array = np.array([[4, 5], [3, 2]])  # pass values to det() function
print(linalg.det(two_d_array))
```

```python
from scipy import linalg
import numpy as np

# define two dimensional array
arr = np.array([[5, 4], [6, 3]])

# pass value into function
eg_val, eg_vect = linalg.eig(arr)  # get eigenvalues and eigenvectors
print(eg_val)  # print eigenvalues
print(eg_vect) # print eigenvectors
```
