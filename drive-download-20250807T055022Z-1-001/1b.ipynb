```python
# Import math Library
import math

# Print the square root of different numbers
print(math.sqrt(10))    # Returns square root as float
print(math.sqrt(12))
print(math.sqrt(68))
print(math.sqrt(100))

# Round square root downward to the nearest integer
print(math.isqrt(10))   # Returns square root as integer
print(math.isqrt(12))
print(math.isqrt(68))
print(math.isqrt(100))
```
